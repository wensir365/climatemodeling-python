# (Associated) Legendre Polynomials

## 1. Legendre polynomials of the 1st kind

```python
%matplotlib inline

import numpy as np
import scipy.special as sp
import pkuplot

x = np.linspace(-1.0, 1.0, 1000)


def Legendre(n,x):
    m = 0
    return sp.lpmv(0,n,x)

#            n
y0 = Legendre(0,x)
y1 = Legendre(1,x)
y2 = Legendre(2,x)
y3 = Legendre(3,x)
y4 = Legendre(4,x)
y5 = Legendre(5,x)

y42 = Legendre(42,x)
y85 = Legendre(85,x)

# plot
pkuplot.plot2d(8,[x,x,x,x,x,x,x,x],[y0,y1,y2,y3,y4,y5,y42,y85],
              style=['k-','c-','g-','b-','r-','y-','m-','m--'],lw=[2,2,2,2,2,2,1,1],
              lb=['$P_0(x)$','$P_1(x)$','$P_2(x)$','$P_3(x)$','$P_4(x)$','$P_5(x)$','$P_{42}(x)$','$P_{85}(x)$'],
              ti='Legendre Polynomials (1st Kind)',xl='x',yl='$P_n(x)$',
              ylim=[-1.1,1.1])
```

## 2. Legendre polynomials of the 2nd kind

### 

```python
#            n
y0 = [sp.lqn(0,nu)[0][-1]   for nu in x]
y1 = [sp.lqn(1,nu)[0][-1]   for nu in x]
y2 = [sp.lqn(2,nu)[0][-1]   for nu in x]
y3 = [sp.lqn(3,nu)[0][-1]   for nu in x]
y4 = [sp.lqn(4,nu)[0][-1]   for nu in x]
y5 = [sp.lqn(5,nu)[0][-1]   for nu in x]

# plot
pkuplot.plot2d(6,[x,x,x,x,x,x],[y0,y1,y2,y3,y4,y5],
              style=['k-','c-','g-','b-','r-','y-'],lw=[2,2,2,2,2,2],
              lb=['$Q_0(x)$','$Q_1(x)$','$Q_2(x)$','$Q_3(x)$','$Q_4(x)$','$Q_5(x)$'],
              ti='Legendre Polynomials (2nd Kind)',xl='x',yl='$Q_n(x)$',
              ylim=[-3,3])
```

## 3. Associated Legendre polynomials of the 1st kind

```python
def AssoLegendre(m,n,x):
    '''
    n-degree(base)
    m-order(shoulder)
     Associated Legendre 
    1 by * sqrt( factorial(n-m)/factorial(n+m) )
    '''
    # Associated Legendre
    y = sp.lpmv(m,n,x)
    # Normalize it
    y = y * np.sqrt(np.math.factorial(n-m)/np.math.factorial(n+m))
    return y
```

### n=0

```python
y00 = AssoLegendre(0,0,x)

pkuplot.plot2d(1,[x,],[y00,],style=['b-',],lw=[4,],lb=['$P_0^0(x)$',],
              ti='Associated Legendre Polynomials (1st Kind)',xl='x',yl='$P_n^m(x)$',
              xlim=[-1,1],ylim=[-1.1,1.1])
```

### n=1

```python
y01 = AssoLegendre(0,1,x)
y11 = AssoLegendre(1,1,x)

pkuplot.plot2d(2,[x,x],[y01,y11],style=['b-','r-'],lw=[4,4],lb=['$P_1^0(x)$','$P_1^1(x)$'],
              ti='Associated Legendre Polynomials (1st Kind)',xl='x',yl='$P_n^m(x)$',
              xlim=[-1,1],ylim=[-1.1,1.1])
```

### n=2

```python
y02 = AssoLegendre(0,2,x)
y12 = AssoLegendre(1,2,x)
y22 = AssoLegendre(2,2,x)

pkuplot.plot2d(3,[x,x,x],[y02,y12,y22],style=['b-','r-','g-'],lw=[4,4,4],
               lb=['$P_2^0(x)$','$P_2^1(x)$','$P_2^2(x)$'],
               ti='Associated Legendre Polynomials (1st Kind)',xl='x',yl='$P_n^m(x)$',
               xlim=[-1,1],ylim=[-1.1,1.1])
```

### n=3

```python
y03 = AssoLegendre(0,3,x)
y13 = AssoLegendre(1,3,x)
y23 = AssoLegendre(2,3,x)
y33 = AssoLegendre(3,3,x)

pkuplot.plot2d(4,[x,x,x,x],[y03,y13,y23,y33],style=['b-','r-','g-','y-'],lw=[4,4,4,4],
               lb=['$P_3^0(x)$','$P_3^1(x)$','$P_3^2(x)$','$P_3^3(x)$'],
               ti='Associated Legendre Polynomials (1st Kind)',xl='x',yl='$P_n^m(x)$',
               xlim=[-1,1],ylim=[-1.1,1.1])
```

### n=4

```python
y04 = AssoLegendre(0,4,x)
y14 = AssoLegendre(1,4,x)
y24 = AssoLegendre(2,4,x)
y34 = AssoLegendre(3,4,x)
y44 = AssoLegendre(4,4,x)

pkuplot.plot2d(5,[x,x,x,x,x],[y04,y14,y24,y34,y44],style=['b-','r-','g-','y-','c-'],lw=[4,4,4,4,4],
               lb=['$P_4^0(x)$','$P_4^1(x)$','$P_4^2(x)$','$P_4^3(x)$','$P_4^4(x)$'],
               ti='Associated Legendre Polynomials (1st Kind)',xl='x',yl='$P_n^m(x)$',
               xlim=[-1,1],ylim=[-1.1,1.1])
```

## 5.  y n=4

```python
ym14 = AssoLegendre(-1,4,x)
ym24 = -AssoLegendre(-2,4,x)
ym34 = AssoLegendre(-3,4,x)
ym44 = -AssoLegendre(-4,4,x)

# P^(0)_4
pkuplot.plot2d(1,[x,],[y04,],style=['k-',],lw=[6,],lb=['$P_4^0(x)$',],
              ti='Associated Legendre Polynomials (1st Kind)',xl='x',yl='$P_n^m(x)$',
              xlim=[-1,1],ylim=[-1.1,1.1])
```
